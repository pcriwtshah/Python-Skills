## What is a vector?

A vector can be described in a few ways. Visually, you could think of a vector as an arrow which is a certain length and pointing in a particular direction. 


````
import matplotlib.pyplot as plt

plt.arrow(0, 0, .5, .5, head_width=.02, head_length=0.01, color = 'y')
plt.arrow(0, 0, .2, .1, head_width=.02, head_length=0.01, color = "r")
plt.arrow(0, 0, .1, .3,  head_width=.02, head_length=0.01, color = "b")
plt.title('Three Vectors')
plt.show()
````

Programatically, we can think of vectors as lists.
For instance, we could represent the 3 vectors above in terms of their coordinates. The coordinates give us directions about how to go from origin (0,0) to the tip of the arrow. The first number indicates how far along the x-axis to travel. The second number indicates how far along the y-axis. This gives you a point and a line is drawn from the origin to the point.

````
yellow = [.5, .5]
red = [.2, .1]
blue = [.1, .3]
````

Mathematically, we show vectors as enclosed in vertical brackets
  \begin{align}
     simpleVector&= \begin{bmatrix}
           x \\
           y
         \end{bmatrix}
  \end{align}

simpleVector is a *2 dimensional* vector, because it has 2 components: x  and y.

Now, what can we do with vectors? 

1) We can add them
\begin{align}
    \begin{bmatrix}
           3 \\
           4
         \end{bmatrix}
         +
            \begin{bmatrix}
           1 \\
           2
         \end{bmatrix}
         =
            \begin{bmatrix}
           4 \\
           6
         \end{bmatrix}
   \end{align}

````
# Let's add our arrows from above
orange = [yellow[0] + red[0], yellow[1] + red[1]]
print("orange", orange)

green =  [yellow[0] + blue[0], yellow[1] + blue[1]]
print('green', green)

purple = [red[0] + blue[0], red[1] + blue[1]]
print('purple', purple)
````

````
# Plot the arrows
plt.arrow(0,0, .5 ,.5, head_width=.02, head_length=0.01, color = 'y')
plt.arrow(0, 0, .2, .1, head_width=.02, head_length=0.01, color = "r")
plt.arrow(0,0, .1, .3,  head_width=.02, head_length=0.01, color = "b")
plt.arrow(0,0, orange[0], orange[1],head_width=.02, head_length=0.01, color ='orange')
plt.arrow(0,0, green[0], green[1],head_width=.02, head_length=0.01, color ='g')
plt.arrow(0,0, purple[0], purple[1],head_width=.02, head_length=0.01, color ='purple')
plt.title("Six Vectors")
plt.show()
````

Notice that the purple and green arrow are on top of each other - this brings us to the second thing that you can do with vectors - you can multiply them by a scalar.

For instance if I multiplied purple by 2 then I would get green.

\begin{align}
             2
            \begin{bmatrix}
           .3 \\
           .4
         \end{bmatrix}
         =
            \begin{bmatrix}
           .6 \\
           .8
         \end{bmatrix}
   \end{align}
   
 In this instance, the 2 is the scalar and you are *scaling* the purple vector

````
# Programatically

green_again = list(map(lambda v: v*2, purple))
print(green_again)
````

You can use scalars to grow, shrink or reverse the direction of a vector.

Let's see it in action on the primary arrows

````
shrink_yellow = list(map(lambda v: v*1/2,yellow))
grow_blue = list(map(lambda v: v*3,blue))
reverse_red = list(map(lambda v: v*-1, red))

plt.arrow(0,0, shrink_yellow[0], shrink_yellow[1], head_width=.02,
          head_length=0.01, color = 'y', alpha = .7)
plt.arrow(0,0, grow_blue[0], grow_blue[1], head_width=.02,
          head_length=0.01, color = 'b', alpha = .7)
plt.arrow(0,0, reverse_red[0], reverse_red[1], head_width=.02,
          head_length=0.01, color = 'r', alpha = .7)
plt.arrow(0, 0, .5 ,.5, head_width=.02, head_length=0.01, color = 'y')
plt.arrow(0, 0, .2, .1, head_width=.02, head_length=0.01, color = "r")
plt.arrow(0, 0, .1, .3,  head_width=.02, head_length=0.01, color = "b")
plt.title("Scaling Vectors")
plt.axis([-.8, 1, -.8, 1])
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
````

Now that we know two things we can do with vectors, add them and scale them, let's talk about a particular type of vector,  *basis vectors*

Basis vectors are the basic building blocks of a space. Through the use of our  2 tools of addition and scaling, we can use basis vectors to "build" any other vector in that space. 

To get a concept of basis vectors, let's stick with 2D space and talk about the classic basis unit vectors of 2D space: ${\hat{i}}$ and ${\hat{j}}$


Question: What 2 variables come to mind when you think of 2 dimensional space? 

````
plt.axis([-1, 1, -1, 1])
plt.axhline(y=0, color='b')
plt.axvline(x=0, color='r')
plt.title("X-axis and Y-axis")
plt.show()
````

Hopefully you thought of X and Y as related to the x-axis and y-axis. Now, how would you represent the x-axis and y-axis as vectors?

1) Choose a point on the x-axis and y-axis


````
plt.axis([-2, 2, -2, 2])
plt.axhline(y=0)
plt.axvline(x=0)
plt.plot(1, 0, marker='o', markersize=10, color="blue")
plt.plot(0, 1, marker='o', markersize=10, color="red")
plt.title("Point on X axis and Y axis")
plt.show()
````

2) What are the coodinates of those points?


````
x = [1,0]
y = [0,1]

# Look 2 vectors! let's plot them as vectors

plt.axis([-2, 2, -2, 2])
#plt.axhline(y=0)
#plt.axvline(x=0)
plt.arrow(0,0, x[0], x[1], head_width=.1, head_length=.25, color = 'b')
plt.arrow(0,0, y[0], y[1],  head_width=.1, head_length=.25, color = 'r')
plt.title('x and y vectors')
plt.show()
````


Let's see what we can do with our x and y vectors. What are the two things we can do with vectors?

Add them and scale them.

````
def add_vectors(c,d):
  new_vector = list(map(lambda a,b: a+b, c, d))
  return new_vector

v_1 = add_vectors(x,y)
v_2 = list(map(lambda v: v*-2, x))
v_3 = add_vectors(v_1, v_2)
v_4 = add_vectors(v_3, v_2)
v_5 = list(map(lambda a: -1*a, y))
v_6 = add_vectors(v_5,v_1)

plt.axis([-2, 2, -2, 2])
plt.arrow(0,0, v_1[0], v_1[1], head_width=.1, head_length=.25, color = "red")
plt.arrow(0,0, v_2[0], v_2[1], head_width=.1, head_length=.25, color = "blue")
plt.arrow(0,0, v_3[0], v_3[1], head_width=.1, head_length=.25, color = "purple")
plt.arrow(0,0, v_4[0], v_4[1],  head_width=.1, head_length=.25, color = "c")
plt.arrow(0,0, v_5[0], v_5[1],  head_width=.1, head_length=.25, color = "yellow")
plt.arrow(0,0, v_6[0], v_6[1],  head_width=.1, head_length=.25, color = "m")

plt.title("all the directions")
plt.show()
````

What I hope you see here is that by using those two vectors and scaling and adding them we can create different vectors that span all of 2 dimensional space.

Does it work with any pair of vectors? 

Nope! The vectors must be *linearly independent* which we can summarize for now as not on top of one another.

## Moving on to Matrices

Here is a possible i-hat and j-hat. 


 $\hat{i} =  \begin{bmatrix}
           1 \\
           0
         \end{bmatrix}\\
         \hat{j}=  \begin{bmatrix}
           0 \\
          2
         \end{bmatrix}$
         
 Now what if instead of representing them as seperate vectors, we put them together in a 2x2 matrix.
$ \begin{bmatrix}
           1 \
           0\\
           0 \
           2
         \end{bmatrix}$
         
  With my 2x2 basis matrix I can multiply any vector with it and have a new transformed vector.
  
  ${\vec{v}} =  \begin{bmatrix}1\ 0\\ 0 \ 2\end{bmatrix}\begin{bmatrix}
                          3\\
                          4
                          \end {bmatrix}  =  
                          \begin{bmatrix}
                          3\\
                          8
                          \end {bmatrix}$
                                         
  That could look a little intimidating at first but notice that it's still scalar multiplication we saw earlier, just in a different format
  
  $3 \begin{bmatrix}
           1 \\
           0
         \end{bmatrix} +  4 \begin{bmatrix}
           0 \\
           2
         \end{bmatrix} = $ 
      
 $ (3 \cdot 1) + (4 \cdot 0) = 3$
 
$ (3 \cdot 0) + (4 \cdot 2) = 8 $

But what if we wanted to multiply two matricies together? 
 $ \begin{bmatrix}
           1 \
           0\\
           0 \
           2
         \end{bmatrix}
         \cdot  \begin{bmatrix}
           3 \
           4\\
           1 \
           6
         \end{bmatrix}$ 
         
 Imagine it broken up into smaller pieces starting with the first column of the 2nd matrix
 
  $ \begin{bmatrix}
           1 \
           0\\
           0 \
           2
         \end{bmatrix} 
         \cdot  \begin{bmatrix}
           3\\
           1\\
           \end {bmatrix} $
           
  Which can be broken down further into
   
   $3 \begin{bmatrix}
           1 \\
           0
         \end{bmatrix} +  1 \begin{bmatrix}
           0 \\
           2
         \end{bmatrix} $
         
   $(3\cdot 1) + (1 \cdot 0) = 3 $
         
   $(3 \cdot 0) + (1 \cdot 2) = 2$
 
 And then the second column of the 2nd matrix
 
   $ \begin{bmatrix}
           1 \
           0\\
           0 \
           2
         \end{bmatrix} 
         \cdot  \begin{bmatrix}
           4\\
           6\\
           \end {bmatrix} $
           
   $4 \begin{bmatrix}
           1 \\
           0
         \end{bmatrix} +
         6 \begin{bmatrix}
           0 \\
           2
         \end{bmatrix} $    
    $(4\cdot 1) + (6 \cdot 0) = 4$  
    $(4 \cdot 0) + (6 \cdot 2) = 12$
         
 So our resulting matrix is:
    $ \begin{bmatrix}
           3  \
           4 \\
           2 \
           12
         \end{bmatrix}$

Let's talk about the determinant of a matrix. The determinant of the matrix is the factor of how it will transform an area in the space... Let's look at some graphs using the matrices from earlier and begin using the NumPy library to illustrate what we're talking about.


````
import numpy as np

# Notice that we input the numbers as rows not columns
m_1 = np.matrix('1 0; 0 2')
print('m_1 \n', m_1)

m_2 = np.matrix('3 4; 1 6')
print('m_2\n', m_2)
````

````
plt.axis([0,12,0,12])
plt.arrow(0,0, m_1[0,0], m_1[1,0], head_width=.1, head_length=.25, color = "red")
plt.arrow(0,0, m_1[0,1], m_1[1,1], head_width=.1, head_length=.25, color = "blue")
plt.fill([0,1,1,0],[0,0,2,2], color = 'purple')


plt.arrow(0,0, m_2[0,0], m_2[1,0], head_width=.1, head_length=.25, color = "orange")
plt.arrow(0,0, m_2[0,1], m_2[1,1], head_width=.1, head_length=.25, color = "c")
plt.fill([0,3,7,4],[0,1,7,6], alpha = .5)
````

The purple matrix has a determinant of 2, meaning that it scales the unit vector (a space of 1 x 1) by a factor of 2.

The blue matrix has a determinant of 14, meaning that it scales the unit vector by a factor of 14.

We calculate the determinant by multiplying the diagonals and subtracting the second diagonal from the first.

For example: 

```
m_2
 [[3 4]
 [1 6]]
 
 (3x6)-(4x1) = 14
```

Let's multiply these matrices together and calculate the determinant



````
# Multiply m_1 and m_2

new_mat = np.matmul(m_1, m_2)
print(new_mat)
````

````
plt.axis([0,15,0,15])
plt.arrow(0,0, m_1[0,0], m_1[1,0], head_width=.1, head_length=.25, color = "red")
plt.arrow(0,0, m_1[0,1], m_1[1,1], head_width=.1, head_length=.25, color = "blue")
plt.fill([0,1,1,0],[0,0,2,2], color = 'purple')

plt.arrow(0,0, m_2[0,0], m_2[1,0], head_width=.1, head_length=.25, color = "orange")
plt.arrow(0,0, m_2[0,1], m_2[1,1], head_width=.1, head_length=.25, color = "c")
plt.fill([0,3,7,4],[0,1,7,6], alpha = .5)

plt.arrow(0,0, new_mat[0,0], new_mat[1,0], head_width=.1, head_length=.25, color = "yellow")
plt.arrow(0,0, new_mat[0,1], new_mat[1,1], head_width=.1, head_length=.25, color = "green")
plt.fill([0,3,7,4],[0,2,14,12], alpha = 0.5)
````

The determinant of the product of m_1 and m_2 is 28! 

What happens when the determinant equals zero?

````
m_3 = np.matrix('2 4; 1 2')
print('determinant of m_3:',np.linalg.det(m_3))

m_4 = np.matmul(m_1, m_3)
print('determinant of m_4:',np.linalg.det(m_4))

plt.axis([0,5,0,5])
# m_3
plt.arrow(0,0, m_3[0,0], m_3[1,0], head_width=.1, head_length=.25, color = "orange")
plt.arrow(0,0, m_3[0,1], m_3[1,1], head_width=.1, head_length=.25, color = "c")
# m_1
plt.arrow(0,0, m_1[0,0], m_1[1,0], head_width=.1, head_length=.25, color = "red")
plt.arrow(0,0, m_1[0,1], m_1[1,1], head_width=.1, head_length=.25, color = "blue")
plt.fill([0,1,1,0],[0,0,2,2], color = 'purple', alpha = .5)
# m_4
plt.arrow(0,0, m_4[0,0], m_4[1,0], head_width=.1, head_length=.25, color = "m")
plt.arrow(0,0, m_4[0,1], m_4[1,1], head_width=.1, head_length=.25, color = "yellow")
````

When a matrix has a determinant of zero then it will squish any other matrix it is multplied with down a dimension. Relatedly, a determinant for a matrix is zero if and only if it has column/row vectors that are linearly dependent - essentially, the matrix doesn't "fill in the shoes" of its dimension (this is called its rank, so another equivalent property is an `n` by `n` matrix with rank less than `n`). And lastly, such a matrix is not invertible.

But when the determinant is non-zero, we can use it to  calculate the *inverse* of a matrix. An inverse of a matrix allows us to *rewind* or *divide* matrix multiplication. Note: the matrix must be square and not all matrices have an inverse. 

To calculate the inverse of a matrix: 

$ \begin{bmatrix}
           a \
           b\\
           c \
           d
         \end{bmatrix}^{-1} = 
         1/determinant \cdot 
         \begin{bmatrix}
           d \
           {-b}\\
           {-c} \
           a
         \end{bmatrix}$
         
 Let's look at this with numbers
 
 $ \begin{bmatrix}
           1 \
           0\\
           0 \
           2
         \end{bmatrix}^{-1} = 
         1/2 \cdot 
         \begin{bmatrix}
           2 \
           {-0}\\
           {-0} \
           1
         \end{bmatrix} =
          \begin{bmatrix}
           1 \
           0\\
           0 \
           .5
         \end{bmatrix}
         $
         
   For an inverse to be valid then the original matrix multiplied by its inverse must equal the *identity matrix*
   
   $ \begin{bmatrix}
           1 \
           0\\
           0 \
           2
         \end{bmatrix} \cdot
           \begin{bmatrix}
           1 \
           0\\
           0 \
           .5
         \end{bmatrix} =
           \begin{bmatrix}
           1 \
           0\\
           0 \
           1
         \end{bmatrix}$
         
 The identity matrix is where the matrix has 1s along the diagonal (going left to right) and zeros everywhere else. 

````
inverse_m_2 = np.linalg.inv(m_2)
print('inverse of m_2:\n',inverse_m_2)
# Does it produce an identity matrix?

print('checking for identity:\n',np.matmul(m_2, inverse_m_2))
````

With numpy the identity can look deceptive. But this is an identity matrix -5.55 x 10^-17 is so small that it would round to zero

So because we have a valid inverse matrix, I can multiply the inverse of m_2 with new_mat and the result will be m_1


````
np.matmul(new_mat, inverse_m_2)
````

The last thing we'll cover tonight is the dot product. You've actually already seen the dot product but we're going to explore it a little more in depth. The dot product allows us to project a vector onto a unit vector

````
x = np.array([1,0])
y = np.array([3,2])
print('Vector y would land at', np.dot(x, y), 'on unit vector x')

plt.axis([-1, 4, -1, 4])
plt.arrow(0,0, x[0], x[1], head_width=.1, head_length=.25, color = "b")
plt.arrow(0,0, y[0], y[1],  head_width=.1, head_length=.25, color = "r")

plt.title('Projection of y onto unit vector x')
plt.show()
````

You can compute the dot product by multiplying "x " by "y-transpose"

A transpose swaps rows to columns or vise versa

````
matrix = np.matrix('1 2 3; 4 5 6 ; 7 8 9')
print('Matrix: \n', matrix)

matrix_transpose = np.transpose(matrix)
print('Matrix transpose: \n', matrix_transpose)
````

For our purposes, the most important aspect of the dot product is "orthogonality"

Two vectors are orthogonal if their dot product equals zero

````
x = np.array([1,1])
y = np.array([2,-2])
print(np.dot(x,y))

plt.axis([-3, 3, -3, 3])
plt.arrow(0,0, x[0], x[1], head_width=.1, head_length=.25, color = "b")
plt.arrow(0,0, y[0], y[1],  head_width=.1, head_length=.25, color = "r")

plt.title("Orthogonal vectors")
plt.show()
````

What is the Data Science application of dot products? They are a fundamental piece of matrix multiplication - each entry in a product matrix is the dot-product of the corresponding row of the first matrix and column of the second matrix.

You saw a bit of matrix multiplication earlier, and you'll see it again tomorrow - it is important in finding the solution to many models.

Note - order matters! Matrix multiplication is *not* commutative (i.e. $M_1 * M_2 != M_2 * M_1$), and the dimension of the matrices involved matters as well ($M_1$ needs the same number of columns as $M_2$ has rows). More on that in future material!
