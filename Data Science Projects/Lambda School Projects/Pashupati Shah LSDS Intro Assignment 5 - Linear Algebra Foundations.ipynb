## Your Answers go here:

1)$ \begin {bmatrix} 1 \\ 3 \\ 4 \end {bmatrix} + \begin {bmatrix} 4 \\ 2 \\ 6 \end {bmatrix} $ =  $ \begin {bmatrix} 1+4 \\ 3+2 \\ 4+6 \end {bmatrix}  $ = $ \begin {bmatrix} 5 \\ 5 \\ 10 \end {bmatrix} $

2)$ 5 \begin {bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} $ = $  \begin {bmatrix} 5 \\ 10 \\ 15 \end{bmatrix} $

3)$ \begin {bmatrix} 1  &  3\\ 2 & 3  \end{bmatrix} \begin {bmatrix} 1 \\ 2 \end{bmatrix} $ = $\begin {bmatrix} 1*1  +  3*2\\ 2*1 + 3*2  \end{bmatrix} $ = $\begin {bmatrix}   7\\ 8 \end{bmatrix} $

4) $ \begin {bmatrix} 1  &  3\\ 2 & 3 \end{bmatrix} \begin {bmatrix} 2 & 2 \\ 3 & 1 \end{bmatrix} $ = $\begin {bmatrix} 1*2  +  3*3 & 1*2 +3*1\\ 2*2 + 3*3 & 2*2 + 3*1  \end{bmatrix} $ = $\begin {bmatrix} 11 & 5\\ 13 & 7\end{bmatrix} $

5) $ Transpose  of \begin{bmatrix} 1 & 4 \\ 2 & 5 \\ 3 & 6 \end{bmatrix} $ = $ \begin{bmatrix} 1 & 2 &3 \\ 4 & 5  & 6 \end{bmatrix} $ 

6)$\begin {bmatrix} 2 & 2 \\ 3 & 1 \end{bmatrix}$ = $(2*1 - 3*2)$ = $(2-6)$ = $-4$

7)  $\begin {bmatrix} 3 & 2 \\ 1 & 2 \end{bmatrix}$ = $1/4 \begin {bmatrix} 3 &1 \\ 2&2 \end{bmatrix}$ = $1$

##Visualize the Vectors

Graph the following:

  Vector v_1 = [1,0] 
  
  Vector v_2 = v_1 scaled by 2
  
  Vector v_3 = [0,-1]
  
  Vector v_4 = the sum of v_2 and v_3
 

```
# Visualize the vectors graph goes here
import matplotlib.pyplot as plt
v_1 = [1, 0]
v_2 = [2, 0]
v_3 = [0, -1]
v_4 = [v_2[0] + v_3[0] , v_2[1] + v_3[1]]
plt.axis([-3, 3, -3, 3])
plt.arrow(0,0, v_1[0], v_1[1], head_width=.1, head_length=.25, color = 'b')
plt.arrow(0,0, v_2[0], v_2[1],  head_width=.1, head_length=.25, color = 'r')
plt.arrow(0,0, v_3[0], v_3[1],  head_width=.1, head_length=.25, color = 'g')
plt.arrow(0,0, v_4[0], v_4[1],  head_width=.1, head_length=.25, color = 'y')
plt.title('x and y vectors')
plt.show()
```

## Graph the Matrix

Graph the columns and rows of this matrix: 
 $\begin {bmatrix} 2 & 2 \\ 3 & 1 \end{bmatrix}$

```
import numpy as np

# Notice that we input the numbers as rows not columns
x_1 = np.matrix('2 2; 3 1')
print('x_1 \n', x_1)
# Matrix graph goes here
plt.axis([0,12,0,12])
plt.arrow(0,0, x_1[0,0], x_1[1,0], head_width=.1, head_length=.25, color = "red")
plt.arrow(0,0, x_1[0,1], x_1[1,1], head_width=.1, head_length=.25, color = "blue")
plt.fill([0,1,1,0],[0,0,2,2], color = 'purple')
```
